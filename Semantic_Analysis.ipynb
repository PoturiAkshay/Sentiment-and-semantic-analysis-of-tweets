```python
#extracting data from reut2-020.sgm file, 
#cleaning the data removing the tags and non alphabet characters
#creating seperate documents for all the articles

import re

InputFile = open("reut2-020.sgm", "r")
counter = -1
count=1
regex = re.compile('[^a-zA-Z0-9]')

for line in InputFile:
    if("<TEXT" in line):
        counter=0
        file=open('./Articles/Article'+str(count)+'.txt','a',encoding='utf-8')
        continue   
    elif("</TEXT>" in line):
        counter=-1 
        count=count+1
        file.close()
        continue
    elif(counter==0):
        line = re.sub("<[^<>]+>", " ", line) 
        tokens=line.split()
        line=""
        for token in tokens:
            new_token=regex.sub('', token)
            if new_token not in (''):
                if new_token.lower()!="reuter":
                    line=line+new_token.lower()+" "
        file.write(line.strip()+"\n")
        continue
             
InputFile.close()
```

```python
#reading all the article files and doing the frequency counts of "Halifax", "Nova Scotia" and "Canada"
#counting the frequency of word "canada" in all the files

import os
import pandas as pd
canada_count=0
halifax_count=0
nova_scotia_count=0

file_tokens=0
canada_in_file=0
file_count=0
df = pd.DataFrame(columns=['term', 'document', 'total_words', 'frequency','f/m'])
index=0

for file in os.listdir("./Articles"):
    if file.endswith(".txt"):
        InputFile = open("./Articles/"+file, "r",encoding='utf-8',)
        tokens = InputFile.read().strip().split()
        file_tokens=file_tokens+len(tokens) 
        
        for word in tokens:
            if ("canada"==word):
                canada_in_file=canada_in_file+1
                #print(tokens)
                
        if('halifax' in tokens):
            halifax_count=halifax_count+1
        if('canada' in tokens):   
            canada_count=canada_count+1
            #print(file)
        if('nova scotia' in tokens):
            nova_scotia_count=nova_scotia_count+1 
            
        if canada_in_file>0:
            df.loc[index] = ("canada",file,file_tokens,canada_in_file,canada_in_file/file_tokens)
        file_tokens=0
        canada_in_file=0
        index=index+1
        InputFile.close()
        file_count=file_count+1
```
